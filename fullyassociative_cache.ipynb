```python
##fully associative cache mapping
def fully_associative_cache(trace, cache_size):
    cache = []
    hit_count = 0
    miss_count = 0

    for address in trace:
        if address in cache:
            # Cache hit
            hit_count += 1
            cache.remove(address)
            cache.append(address)
        else:
            # Cache miss
            miss_count += 1
            if len(cache) >= cache_size:
                cache.pop(0)
            cache.append(address)

    return hit_count, miss_count


def simulate_cache(trace, cache_size):
    hits, misses = fully_associative_cache(trace, cache_size)
    total_accesses = hits + misses
    hit_rate = hits / total_accesses

    print(f"Total Accesses: {total_accesses}")
    print(f"Hit Rate: {hit_rate:.2%}")
    print(f"{'Access':>8} {'Result':>8} {'Cache Status':>20}")

    cache = []

    for i, address in enumerate(trace):
        if address in cache:
            result = "Hit"
            cache.remove(address)
            cache.append(address)
        else:
            result = "Miss"
            if len(cache) >= cache_size:
                cache.pop(0)
            cache.append(address)

        # Display the current status of the cache
        print(f"{address:#010x} {result:>8} {''.join(f'{x:#010x} ' for x in cache)}")


# Example memory access trace
trace_file = [
    0x80000000, 0x80000004, 0x80000008, 0x8000000C,
    0x00000020, 0x80000010, 0x80000014, 0x80000018,
    0x8000000C, 0x00000024, 0x80000010, 0x80000014,
    0x80000018, 0x8000000C, 0x00000028, 0x80000010,
    0x80000014, 0x80000018, 0x8000000C, 0x0000002C,
    0x80000010, 0x80000014, 0x80000018, 0x8000000C,
    0x00000030, 0x80000010, 0x80000014, 0x80000018,
    0x8000000C, 0x00000034, 0x80000010, 0x80000014,
    0x80000018, 0x8000000C, 0x00000038, 0x80000010,
    0x80000014, 0x80000018, 0x8000000C, 0x0000003C,
    0x80000010, 0x80000014, 0x80000018, 0x8000000C,
    0x00000040, 0x80000010, 0x80000014, 0x80000018,
    0x8000000C, 0x00000044, 0x80000010, 0x80000014,
    0x80000018, 0x8000000C, 0x00000048, 0x80000010,
    0x80000014, 0x80000018, 0x8000000C, 0x0000004C,
    0x80000010, 0x80000014, 0x80000018, 0x8000000C,
    0x00000050, 0x80000010, 0x80000014, 0x80000018,
    0x8000000C, 0x00000054, 0x80000010, 0x80000014,
    0x80000018, 0x8000000C, 0x00000058, 0x80000010,
    0x80000014, 0x80000018, 0x8000000C, 0x0000005C,
    0x80000010, 0x80000014, 0x80000018, 0x8000000C,
    0x00000060, 0x80000010, 0x80000014, 0x80000018,
    0x8000000C, 0x00000064, 0x80000010, 0x80000014,
    0x80000018, 0x8000000C, 0x00000068, 0x80000010,
    0x80000014, 0x80000018, 0x8000000C, 0x0000006C,
    0x80000010, 0x80000014, 0x80000018
]

# Set cache size
cache_size = 8

# Simulate the cache
simulate_cache(trace_file, cache_size)
```
